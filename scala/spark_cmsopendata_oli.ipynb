```scala
//
// Convert Dataset[Row] => Dataset[Event]
//
val ds_muons = df_muons.as[Event]
```

```scala
//
// Let's try a few simple queries: Filter out muons with pt<20
// 
import scala.math._
val ds_muons_filtered = ds_muons.map({case event => 
    event.recoMuons_muons__RECO_obj.filter({case muon => muon.pt_ > 20 && abs(muon.eta_) < 2.4})
})
```

```scala
//
// Select only pts
//
val ds_pts = ds_muons_filtered.select("value.pt_")
```
